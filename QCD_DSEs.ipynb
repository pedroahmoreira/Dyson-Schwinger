## Equações de Dyson-Schwinger na Cromodinâmica Quântica

A teoria em foco agora é a Cromodinâmica quântica (QCD), uma teoria de calibre com simetria $SU(3)$ que descreve as interações fortes. Sua densidade lagrangiana completa é
\begin{equation}
    \mathcal{L}_{QCD} = \bar{q}_{ij} (i \gamma^{\mu} (D_{\mu})_{ij} - m \delta_{ij}) q_{ij} - \frac{1}{4} G^{\mu \nu}_a G^a_{\mu \nu} + \frac{1}{2} \eta^*_a(\delta_{ab} {\Box}^2 + gf_{abc} G^\mu_c \partial_\mu)\eta_a + \frac{1}{2\alpha} (\partial^\mu G_\mu)(\partial^\nu G_\nu),
\end{equation}
em que $D_{\mu} = \partial_\mu + igG_\mu$ é a derivada covariante, $q$ são campos de quarks, $G_\mu$ são campos de gluons e $\eta_a$ são campos fantasmas. 

Para nossas deduções, podemos desconsiderar os outros termos e ignorar os índices $i,j,a$:
\begin{equation}
    \mathcal{L}_{QCD} = \bar{q} (i \gamma^{\mu} D_{\mu} - m) q - \frac{1}{4} G^{\mu \nu} G_{\mu \nu}.
\end{equation}

A ação é, portanto,
\begin{equation}
    S[q, \bar{q}, G_\mu] = \int d^4x \ \mathcal{L}_{QCD}.
\end{equation}

### Derivação das DSEs

Como anteriormente, temos como ponto de partida a relação
\begin{equation}
    \int^{+\infty}_{-\infty} dx \ \frac{\partial \rho}{\partial x} = 0 \ \Leftrightarrow \ \rho (\pm \infty) = 0,
    \tag{$\star$}
\end{equation}
que é simplesmente uma aplicação do teorema fundamental do cálculo.

Especificamente, vamos utilizar as três identidades que são casos particulares de $(\star)$:
\begin{align}
    &\int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ \frac{\delta}{\delta q} \ e^{-S + \bar{q}K + \bar{K}q + J_\mu G_\mu} = 0 \tag{$*$},\\ 
    &\int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ \frac{\delta}{\delta G_\nu} \ e^{-S + \bar{q}K + \bar{K}q + J_\mu G_\mu} = 0 \tag{$**$},\\ 
    &\int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ \frac{\delta}{\delta \eta^*} \ e^{-S + \bar{q}K + \bar{K}q + J_\mu G_\mu} = 0 \tag{$***$},
\end{align}
em que $q,G_\mu,\eta \rightarrow 0$ no infinito.

Também será útil a identidade funcional
\begin{equation}
    \int \mathcal{D} \phi \ \phi(x) \ e^{-S[\phi] + J(x) \cdot \phi(x)} = \frac{\delta }{\delta J(x)} \int \mathcal{D} \phi \ e^{-S[\phi] + J(x) \cdot \phi(x)}.
    \tag{$\star \star$}
\end{equation}

### DSE para Propagador de Quarks

Começamos com a identidade ($*$):
\begin{equation*}
    \int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ \frac{\delta}{\delta q} \ e^{-S + \bar{q}K + \bar{K}q + J_\mu G_\mu} = 0.
\end{equation*}

Avaliando o lado esquerdo da igualdade (atuando a derivada funcional), temos
\begin{equation*}
    \int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ \left( - \frac{\delta S}{\delta q} + \bar{K} \right) \ e^{-S + \bar{q}K + \bar{K}q + J_\mu G_\mu} = 0
\end{equation*}
que podemos mover para fora da integral pela identidade $(\star \star)$  como
\begin{equation*}
    \left[- \frac{\delta S}{\delta q} \left(\frac{\delta}{\delta \bar{K}}, \frac{\delta}{\delta K}, \frac{\delta}{\delta J} \right) + \ \bar{K} \ \right] \underbrace{\int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ e^{-S + \bar{q}K + \bar{K}q + J_\mu G_\mu}}_{Z[\bar{K}, K, J]} = 0.
\end{equation*}

Portanto, temos a equação de Dyson-Schwinger, em forma reduzida, no espaço da posição $(x = x^\mu)$:
\begin{equation}
    \left[ \frac{\delta S}{\delta q(x)} \left(\frac{\delta}{\delta \bar{K}(x)}, \frac{\delta}{\delta K(x)}, \frac{\delta}{\delta J(x)} \right) - \ \bar{K}(x) \ \right] Z[\bar{K}, K, J] = 0.
\end{equation}

Podemos substituir a ação dada pela lagrangiana $\mathcal{L}_{QCD}$:
\begin{align*}
    \frac{\delta S}{\delta q(x)} = \frac{\delta}{\delta q(x)} \int d^4y \left\{ \bar{q}(y) [i \gamma^{\mu} D_{\mu} - m] q(y) - \frac{1}{4} G^{\mu \nu} G_{\mu \nu} \right\} &= \int d^4y \ \bar{q}(y) [i \gamma^{\mu} D_{\mu} - m] \delta^4(x - y) \\
    &= \ \bar{q}(x) [i \gamma^{\mu} \partial_{\mu} - m] - g \bar{q}(x) \gamma^\mu G_\mu
\end{align*}
portanto
\begin{equation*}
    \left[ \frac{\delta S}{\delta q} \left(\frac{\delta}{\delta \bar{K}}, \frac{\delta}{\delta K}, \frac{\delta}{\delta J} \right) - \ \bar{K} \ \right] Z = \left[ (i \partial\!\!\!/_x - m) \frac{\delta}{\delta K} - g \gamma^\mu \frac{\delta}{\delta K} \frac{\delta}{\delta J^\mu} - \bar{K} \right] Z = 0,
\end{equation*}
em que usamos a notação _slash_ de Feynman: $\partial\!\!\!/_x = \gamma^\mu \partial_\mu$, com a derivada em relação a $x^\mu$.

\end{equation}

#### Passando para o Espaço do Momento

O propagador para um férmion livre no espaço da posição é dado por
\begin{equation}
    (i \partial\!\!\!/_x - m) S_0(x - y) = \delta^4(x - y) \tag{$i$},
\end{equation}
e no espaço do momento é
\begin{equation}
    (p\!\!\!/ - m) S_0(p) = 1. \tag{$ii$}
\end{equation}

A transformada de Fourier que nos leva ao espaço de momento é dada por
\begin{equation}
    A(p) = \int d^4x \ e^{ip \cdot x} A(x) \tag{$j$}
\end{equation}
e sua inversa é
\begin{equation}
    A(x) = \int \frac{d^4p}{(2\pi)^4} \ e^{- ip \cdot x} A(p). \tag{$jj$}
\end{equation}

Assim, podemos fazer a transformada da equação $(4)$:
\begin{equation*}
    e^{ip \cdot x} \left[ \frac{\delta S}{\delta q(x)} \left(\frac{\delta}{\delta \bar{K}(x)}, \frac{\delta}{\delta K(x)}, \frac{\delta}{\delta J(x)} \right) - \ \bar{K}(x) \ \right] Z = 0,
\end{equation*}
portanto
\begin{equation*}
    \int d^4x \ e^{ip \cdot x} \left[- \frac{\delta S}{\delta q(x)} \left(\frac{\delta}{\delta \bar{K}(x)}, \frac{\delta}{\delta K(x)}, \frac{\delta}{\delta J(x)} \right) - \ \bar{K}(x) \ \right] Z = 0,
\end{equation*}
que nos dá
\begin{equation}
    \left[ \frac{\delta S}{\delta q(p)} \left((2\pi)^4 \frac{\delta}{\delta \bar{K}(-p)}, (2\pi)^4 \frac{\delta}{\delta K(-p)}, (2\pi)^4 \frac{\delta}{\delta J(-p)} \right) - \ \bar{K}(p) \ \right] Z[\bar{K}, K, J] = 0
\end{equation}
que é a forma reduzida da DSE no spaço de momento.

Aqui foi usado o fato de que, como
\begin{equation*}
    \phi(x) = \int \frac{d^4p}{(2\pi)^4} \ e^{- ip \cdot x} \phi(p), \ \ \ \ \ \ \ J(x) = \int \frac{d^4q}{(2\pi)^4} \ e^{- iq \cdot x} J(q),
\end{equation*}
pela equação $(jj)$, então no termo de fase do funcional gerador teremos
\begin{align*}
    \int d^4x \ J(x) \phi(x) =  \int d^4x \int \frac{d^4q}{(2\pi)^4} \ e^{- ip \cdot x} J(q) \int \frac{d^4p}{(2\pi)^4} \ e^{- iq \cdot x} \phi(p) &= \iint \frac{d^4q}{(2\pi)^4} \frac{d^4p}{(2\pi)^4} \ J(q) \phi(p) \int d^4x \ e^{-i(p+q) \cdot x} \\

\end{align*}

\begin{equation}
    \int d^4x \ J(x) \phi(x) = \int \frac{d^4p}{(2\pi)^4} \ J(-p) \phi(p).
\end{equation}
