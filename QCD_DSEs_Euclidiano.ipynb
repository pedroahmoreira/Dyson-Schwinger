# Equações de Dyson-Schwinger da Cromodinâmica Quântica no espaço euclidiano

A teoria em foco é a Cromodinâmica quântica (QCD), uma teoria de calibre com simetria $SU(3)$ que descreve as interações fortes. Sua densidade lagrangiana completa, no espaço de coordendadas, é
\begin{equation}
    \mathcal{L}_{QCD} = \bar{q}_{ij} (i \gamma^{\mu} (D_{\mu})_{ij} - m \delta_{ij}) q_{ij} - \frac{1}{4} G^{\mu \nu}_a G^a_{\mu \nu} + \frac{1}{2} \eta^*_a(\delta_{ab} {\Box}^2 + gf_{abc} G^\mu_c \partial_\mu)\eta_a + \frac{1}{2\alpha} (\partial^\mu G_\mu)(\partial^\nu G_\nu),
\end{equation}
em que $D_{\mu} = \partial_\mu + igG_\mu$ é a derivada covariante, $q$ são campos de quarks, $G_\mu$ são campos de gluons e $\eta_a$ são campos fantasmas. 

Para nossas deduções, vamos trabalhar com a métrica euclidiana
\begin{align*}
    x_0 &= -ix_4 \\
    \partial_0 &= i\partial_4 \\
    \gamma_0 &= \gamma_4 \\
    \gamma_i &= -i\gamma_i
\end{align*}

e transformar a lagrangiana para o espaço do momento, ignorando os índices $i,j$. Com essas duas mudanças obtemos:
\begin{align}
    \mathcal{L}_{QCD} &= \bar{q} S^{-1}_0(p)q + ig\bar{q}\gamma^\mu G_{a\mu} q + \frac{1}{2} G_{a\mu} (D^{\mu\nu}_0)^{-1} G_{a\nu} - \frac{1}{2} \eta^*_a p^2 \eta_a \nonumber \\
    &+ \frac{1}{4} g^2 (f_{abc}G_{b\mu}G_{c\nu})(f_{ade}G_{d\mu}G_{e\nu}) - \frac{i}{2}g(f_{abc}G_{b\mu}G_{c\nu})(p_\mu G_{a\nu} - p_\nu G_{a\mu}) + \frac{i}{2}gf_{abc}G_{c\mu} \eta^*_a p_\mu \eta_b
\end{align}

A ação da QCD é, portanto,
\begin{equation}
    \mathcal S[q, \bar{q}, G] = \int d^4p \ \mathcal{L}_{QCD},
\end{equation}
e o funcional gerador da teoria quântica de campos é
\begin{equation}
    Z[\bar{K}, K, J] = \int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ e^{-\mathcal{S} + \bar{q}K + \bar{K}q + J_\mu G^\mu},
\end{equation}
em que algo do tipo $J\phi$ denota o produto interno $J(p) \cdot \phi(p) = \int d^4p \ J(p) \phi(p)$.

## Derivação das DSEs

Temos como ponto de partida a relação
\begin{equation}
    \int^{+\infty}_{-\infty} dx \ \frac{\partial \rho}{\partial x} = 0 \ \Leftrightarrow \ \rho (\pm \infty) = 0,
    \tag{$\star$}
\end{equation}
que é simplesmente uma aplicação do teorema fundamental do cálculo.

Especificamente, vamos utilizar as três identidades que são casos particulares de $(\star)$:

\begin{align}
    &\int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ \frac{\delta}{\delta q} \ e^{-\mathcal{S} + \bar{q}K + \bar{K}q + J_\mu G_\mu} = 0 \tag{$*$},\\

    &\int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ \frac{\delta}{\delta G_\nu} \ e^{-\mathcal{S} + \bar{q}K + \bar{K}q + J_\mu G_\mu} = 0 \tag{$**$},\\

    &\int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ \frac{\delta}{\delta \eta^*} \ e^{-\mathcal{S} + \bar{q}K + \bar{K}q + J_\mu G_\mu} = 0 \tag{$***$},
\end{align}
em que $q,G_\mu,\eta \rightarrow 0$ no infinito.

Também será útil a identidade funcional
\begin{equation}
    \int \mathcal{D} \phi \ \phi(x) \ e^{-\mathcal{S}[\phi] + J(x) \cdot \phi(x)} = \frac{\delta }{\delta J(x)} \int \mathcal{D} \phi \ e^{-\mathcal{S}[\phi] + J(x) \cdot \phi(x)}.
    \tag{$\star \star$}
\end{equation}

### DSE para o Propagador de Quarks

Começamos com a identidade $(*)$:
\begin{equation*}
    \int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ \frac{\delta}{\delta q} \ e^{-\mathcal{S} + \bar{q}K + \bar{K}q + J_\mu G_\mu} = 0.
\end{equation*}

Avaliando o lado esquerdo da igualdade (atuando a derivada funcional), temos
\begin{equation*}
    \int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ \left( - \frac{\delta \mathcal{S}}{\delta q} + \bar{K} \right) \ e^{-\mathcal{S} + \bar{q}K + \bar{K}q + J_\mu G_\mu} = 0
\end{equation*}
que podemos mover para fora da integral pela identidade $(\star \star)$  como
\begin{equation*}
    \left[- \frac{\delta \mathcal{S}}{\delta q} \left(\frac{\delta}{\delta \bar{K}}, \frac{\delta}{\delta K}, \frac{\delta}{\delta J} \right) + \bar{K} \right] \underbrace{\int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ e^{-\mathcal{S} + \bar{q}K + \bar{K}q + J_\mu G_\mu}}_{Z[\bar{K}, K, J]} = 0.
\end{equation*}

Portanto, temos a equação de Dyson-Schwinger, em forma reduzida, no espaço do momento:
\begin{equation}
    \left[ \frac{\delta \mathcal{S}}{\delta q(p)} \left(\frac{\delta}{\delta \bar{K}(p)}, \frac{\delta}{\delta K(p)}, \frac{\delta}{\delta J(p)} \right) - \bar{K}(p) \right] Z[\bar{K}, K, J] = 0.
\end{equation}

\end{equation*}


Podemos substituir a ação, dada pela lagrangiana $\mathcal{L}_{QCD}$, cuja derivada é:

\begin{align*}
    \frac{\delta \mathcal{S}}{\delta q(p)} &= \frac{\delta}{\delta q(p)} \int d^4k \left\{ \bar{q} S^{-1}_0(k)q + ig\bar{q}\gamma^\mu G_{a\mu} q + \frac{1}{2} G_{a\mu} (D^{\mu\nu}_0)^{-1} G_{a\nu} - \frac{1}{2} \eta^*_a p^2 \eta_a + ...\right\} \\

    &= \int d^4k \ \left\{ \bar{q} S^{-1}_0(k) \delta^4(p - k) + ig\bar{q}\gamma^\mu G_{a\mu}(k) \delta^4(p - k) \right\} \\

    &= \ \bar{q} S^{-1}_0(p) + ig\bar{q}\gamma^\mu G_{a\mu}(p)
\end{align*}
portanto
\begin{equation*}
    \left[ \frac{\delta \mathcal{S}}{\delta q} \left(\frac{\delta}{\delta \bar{K}}, \frac{\delta}{\delta K}, \frac{\delta}{\delta J} \right) - \bar{K} \right] Z = \left[ S^{-1}_0 \frac{\delta}{\delta K} + ig \gamma^\mu \frac{\delta}{\delta K} \frac{\delta}{\delta J^\mu} - \bar{K} \right] Z = 0.
\end{equation*}

Assim, ficamos com
\begin{equation}
    \left[ S^{-1}_0(p) \frac{\delta}{\delta K(p)} + ig \gamma^\mu \frac{\delta}{\delta K(p)} \frac{\delta}{\delta J^\mu (p)} - \bar{K}(p) \right] Z[\bar{K}, K, J] = 0.
\end{equation}

\begin{equation*}

\end{equation*}

Agora podemos reorganizar e fazer uma derivada adicional em relação a $\bar{K}(q)$:

\begin{align*}
    \frac{\delta}{\delta \bar{K}(q)} \left[ S^{-1}_0(p) \frac{\delta}{\delta K(p)} + ig \gamma^\mu \frac{\delta}{\delta K(p)} \frac{\delta}{\delta J^\mu(p)} \right] Z &= \frac{\delta}{\delta \bar{K}(q)} \bar{K}(p)Z \\ \therefore

    \left[ S^{-1}_0(p) \frac{\delta^2}{\delta K(p) \delta \bar{K}(q)} + ig \gamma^\mu \frac{\delta}{\delta J^\mu(p)} \frac{\delta^2}{\delta K(p) \delta \bar{K}(q)} \right] Z &= \delta^4(q - p)Z + \bar{K}(p) \frac{\delta Z}{\delta \bar{K}(q)}.
\end{align*}

Identificamos aqui a função de dois pontos dos quarks:
\begin{equation}
    S(p, q) = Z^{-1} \frac{\delta^2}{\delta K(p) \delta \bar{K}(q)} \tag{$i$}.
\end{equation}

Substituindo, obtemos
\begin{equation*}
    \left[ S^{-1}_0(p) + ig \gamma^\mu \frac{\delta}{\delta J^\mu(p)} \right] Z S(p, q) = \delta^4(q - p)Z + \bar{K}(p) \frac{\delta Z}{\delta \bar{K}(q)}
\end{equation*}
que podemos expandir usando a regra do produto:
\begin{equation*}
    \left[ Z S^{-1}_0(p) + ig \gamma^\mu \frac{\delta Z}{\delta J^\mu(p)} + ig Z \gamma^\mu \frac{\delta}{\delta J^\mu(p)} \right] S(p, q) = \delta^4(q - p)Z + \bar{K}(p) \frac{\delta Z}{\delta \bar{K}(q)}.
\end{equation*}

Dividindo por Z:
\begin{equation}
    \left[ S^{-1}_0(p) + ig Z^{-1} \gamma^\mu \frac{\delta Z}{\delta J^\mu(p)} + ig \gamma^\mu \frac{\delta}{\delta J^\mu(p)} \right] S(p, q) = \delta^4(q - p) + \bar{K}(p) Z^{-1} \frac{\delta Z}{\delta \bar{K}(q)}
\end{equation}
em que identificamos o segundo termo do lado esquerdo da equação como o valor esperado do campo de gluons, que deve ser nulo quando a fonte $J^\mu$ vai para zero.

Fazendo $J^\mu, \bar{K} \rightarrow 0$, obtemos
\begin{equation}
    \left[ S^{-1}_0(p) + ig \gamma^\mu \frac{\delta}{\delta J^\mu(p)} \right] S(p, q) = \delta^4(q - p),
\end{equation}
que é a equação de Dyson-Schwinger para o propagador de quarks.

### DSE para o propagador de Glúons

Desta vez, começamos com a identidade $(**)$:
\begin{equation*}
    \int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ \frac{\delta}{\delta G_\nu} \ e^{-\mathcal{S} + \bar{q}K + \bar{K}q + J_\mu G_\mu} = 0
\end{equation*}

Avaliando a derivada, temos
\begin{align*}
    \int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ \left( - \frac{\delta \mathcal{S}}{\delta G_\nu} + J_\mu \right) \ e^{-\mathcal{S} + \bar{q}K + \bar{K}q + J_\mu G_\mu} &= 0 \\
    \therefore \left[- \frac{\delta \mathcal{S}}{\delta G_\nu} \left(\frac{\delta}{\delta \bar{K}}, \frac{\delta}{\delta K}, \frac{\delta}{\delta J} \right) + J_\mu \right] \underbrace{\int \mathcal{D}(q, \bar{q}, G, \eta, \eta^*) \ e^{-\mathcal{S} + \bar{q}K + \bar{K}q + J_\mu G_\mu}}_{Z[\bar{K}, K, J]} &= 0.
\end{align*}

Logo, a DSE em forma reduzida é
\begin{equation}
    \left[\frac{\delta \mathcal{S}}{\delta G_\nu (p)} \left(\frac{\delta}{\delta \bar{K}(p)}, \frac{\delta}{\delta K(p)}, \frac{\delta}{\delta J(p)} \right) - J_\mu (p) \right] Z[\bar{K}, K, J]= 0.
\end{equation}

    &+ \left. \frac{1}{4} g^2 (f_{abc}G_{b\lambda}G_{c\sigma})(f_{ade}G_{d\lambda}G_{e\sigma}) - \frac{i}{2}g(f_{abc}G_{b\lambda}G_{c\sigma})(p_\lambda G_{a\sigma} - p_\sigma G_{a\lambda}) + \frac{i}{2}gf_{abc}G_{c\lambda} \eta^*_a p_\lambda \eta_b \right\}

### Vértice de interação Quark-Glúon

Podemos expandir o segundo termo da equação para o propagador de quarks, usando a regra da cadeia:
\begin{equation}
    \frac{\delta S(p, q)}{\delta J_\mu (p)} = \int_k \frac{\delta \langle G_\nu(k) \rangle}{\delta J_\mu (p)} \frac{\delta S(p, q)}{\delta \langle G_\nu(k) \rangle} \tag{$\clubsuit$}
\end{equation}
